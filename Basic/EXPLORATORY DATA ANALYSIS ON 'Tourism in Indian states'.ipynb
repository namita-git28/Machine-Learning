#### DATA VISUALIZATION

```python
states = df['State/UT'].values
domestic_growth_rate = df['Growth Rate - Domestic'].values
foreign_growth_rate = df['Growth Rate - Foreign'].values
domestic_2016 = df['2016 - Domestic'].values
foreign_2016 = df['2016 - Foreign'].values
domestic_2017 = df['2017 - Domestic'].values
foreign_2017 = df['2017 - Foreign'].values
```

```python
#To know the correlation between different attributes
correlation = df.corr()
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns,annot=True)
```

##### Here -ve value represents -ve correlation & +ve value represents +ve correlation between the attributes.

```python
plt.figure(figsize=(10,10))
plt.bar(states,domestic_growth_rate,color='blue',edgecolor='red')
plt.tick_params(axis='x', which='major', labelsize=10)
plt.xticks(rotation=90)
plt.ylabel('Growth')
plt.xlabel('States/UT')
plt.title("Domestic Growth Across States 2016-17(in %)",fontdict={'fontsize':15,'fontweight':5})
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.bar(states,foreign_growth_rate,color='yellow',edgecolor='green')
plt.tick_params(axis='x', which='major', labelsize=10)
plt.xticks(rotation=90)
plt.ylabel('Growth'),plt.xlabel('States/UT')
plt.title("Foreign Growth Across States 2016-17(in %)",fontdict={'fontsize':15,'fontweight':5})
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(states,domestic_2016,color='green'),plt.tick_params(axis='x', which='major', labelsize=10)
plt.xticks(rotation=90)
plt.ylabel('Domestic tourists in 2016')
plt.xlabel('States/UT')
plt.title("Domestic Tourists Across States 2016 (in le8)",fontdict={'fontsize':12,'fontweight':5})
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(states,foreign_2016,color='red')
plt.tick_params(axis='x', which='major', labelsize=10)
plt.xticks(rotation=90)
plt.ylabel('Foreign tourists in 2016')
plt.xlabel('States/UT')
plt.title("Foreign Tourists Across States 2016 (in le6)",fontdict={'fontsize':12,'fontweight':5})
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(states,domestic_2017,color='orange')
plt.tick_params(axis='x', which='major', labelsize=10)
plt.xticks(rotation=90)
plt.ylabel('Domestic tourists in 2017')
plt.xlabel('States/UT')
plt.title("Domestic Tourists Across States 2017 (in le8)",fontdict={'fontsize':12,'fontweight':5})
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(states,foreign_2017,color='pink')
plt.tick_params(axis='x', which='major', labelsize=10)
plt.xticks(rotation=90)
plt.ylabel('Foreign tourists in 2017')
plt.xlabel('States/UT')
plt.title("Foreign Tourists Across States 2017 (in le6)",fontdict={'fontsize':12,'fontweight':5})
plt.show()
```
